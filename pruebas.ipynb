```python
import pandas as pd
import re
```

```python
def info_dataframe(archivo):
    patron_extension= "\..*"
    extension = re.findall(patron_extension, archivo)
    if extension == ".csv":
        csv= pd.read_csv(archivo)
    return csv
```
